## 過濾掉盤後交易&到期月分不符格式者

```python
filter1 = (df['交易時段']!='盤後')
filter2 = (df['到期月份(週別)'].apply(lambda x:len(x)<=6))
df2 = df[(filter1 & filter2)].copy()
df2
```

```python
df_dt = pd.to_datetime(df2.交易日期,format="%Y/%m/%d").copy()
df2["交易日期"] = pd.DataFrame({"交易日期":df_dt}).copy()
df2
```

```python
df2 = df2.drop(columns = ['契約'])
df2
```

```python
df2['到期月份(週別)']=df2['到期月份(週別)'].astype('str')
df2['交易日期']=df2['交易日期'].astype('str')
df2['結算價']=df2['結算價'].astype('float')
df2.info()
```

```python
df2['交易日期'] = df2['交易日期'].apply(lambda x:x.split("-")[1]).copy()
```

```python
df2['交易日期']=df2['交易日期'].astype('int')
```

```python
df2['到期月份(週別)'] = df2['到期月份(週別)'].apply(lambda x:x[4:6]).copy()
```

```python
df2
```

```python
df2['到期月份(週別)'] = df2['到期月份(週別)'].astype('int')
```

```python
filter4 = (df2['交易日期']+2 == df2['到期月份(週別)'])
filter5 = (df2['交易日期']-10 == df2['到期月份(週別)'])
df3 = df2[(filter4 | filter5)]

```

```python
df3
```

```python
df3.set_index(np.arange(len(df3['交易日期'])))
```

```python
df3.groupby(['交易日期','到期月份(週別)']).mean().to_csv('20210316練習成果.csv',encoding='big5')
```
