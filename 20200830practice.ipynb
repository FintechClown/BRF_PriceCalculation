```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\USER\Desktop\pypractice\2018_fut.csv',encoding = 'big5')
df
```

```python
filter1 = (df['交易時段']!='盤後')
filter2 = (df['到期月份(週別)'].apply(lambda x:len(x)<=6))
df2 = df[(filter1 & filter2)]
df2
```

```python
df_dt = pd.to_datetime(df2.交易日期,format="%Y/%m/%d")
df2['交易日期'] = df_dt
df2
```

```python
df.drop(['契約'],axis=1,inplace=True)
df
```

```python
df2
```

```python
df2['交易日期']=df2['交易日期'].apply(lambda x :x.split('/')[1])
```

```python
df3 = df2.rename(columns = { '交易日期' : '交易月份'}).copy()
df3
```

```python
df3['到期年'] = df3['到期月份(週別)'].apply(lambda x :x[0:4]).copy()
df3['到期月份(週別)']=df3['到期月份(週別)'].apply(lambda x :x[4:6]).copy()

```

```python
df3['到期月份(週別)']=df3['到期月份(週別)'].astype('int')
df3['交易月份']=df3['交易月份'].astype('int')
df3['結算價']=df3['結算價'].astype('float')
df3.info()
```

```python
df3
```

```python
filter4 = (df3['交易月份']+2 == df3['到期月份(週別)'])
filter5 = (df3['交易月份']-10 == df3['到期月份(週別)'])
df4 = df3[(filter4 | filter5)]

```

```python
df4.set_index(np.arange(len(df4['交易月份'])))
```

```python
df4.groupby(['交易月份','到期月份(週別)']).mean().to_csv('20200830練習成果.csv',encoding='big5')
```
